```python
from jinja2 import Environment, FileSystemLoader
import json
```

```python
def page_link(a):
    if ' ' in a:
        return a.replace(" ", "_")
    else: 
        return a
```

```python
environment = Environment(loader=FileSystemLoader("templates/"))
environment.globals['page_link'] = page_link
```

```python
f = open('jsons/general.json')
general = json.load(f)

e = open('jsons/homepage.json')
Homepage = json.load(e)
```

```python
template = environment.get_template("homepage.html.j2")
```

```python
content = template.render(
        general=general,
        Homepage=Homepage
    )
```

```python
with open("Home.html", mode="w", encoding="utf-8") as Homepage:
        Homepage.write(content)
```

```python
f.close()
e.close()
```
