```python
from jinja2 import Environment, FileSystemLoader
import json
```

```python
def page_link(a):
    if ' ' in a:
        return a.replace(" ", "_")
    else: 
        return a
```

```python
environment = Environment(loader=FileSystemLoader("templates/"))
environment.globals['page_link'] = page_link
```

```python
template = environment.get_template("research.html.j2")
```

```python
json_files = ["general", "research"]
data = {}
for json_file in json_files:
    with open(f"jsons/{json_file}.json") as json_var:
        data[json_file] = json.load(json_var)
```

```python
content = template.render(general=data["general"], research=data["research"])
```

```python
with open("Research.html", mode="w", encoding="utf-8") as research:
        research.write(content)
```
